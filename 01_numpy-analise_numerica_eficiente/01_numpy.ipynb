### Como calcular o coeficiente angular (a)

a = Coeficiente angular <br>
n = número de elementos <br>
Y = Moscow <br>
X = Datas

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
print('Coeficiente angular')
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X ** 2) - np.sum(X) ** 2)
a
```

### Coeficiente linear

```python
b = np.mean(Y) - a * np.mean(X)
b
```

```python
y = a * X + b
```

```python
print('Processo de regressão linear')
np.linalg.norm(Moscow - y).round(3)
```
